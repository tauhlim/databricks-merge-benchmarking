```python
dbutils.widgets.text("catalog", "field_demos", "Catalog to use")
dbutils.widgets.text("schema", "tauherng", "Schema to use")
dbutils.widgets.text("clone_data", "Yes", "Whether to reset data by re-cloning")
```

```python
CATALOG = dbutils.widgets.get("catalog")
SCHEMA = dbutils.widgets.get("schema")
```

```python
%sql
CREATE SCHEMA IF NOT EXISTS IDENTIFIER(CONCAT(:catalog,".", :schema,"_classic"));
CREATE SCHEMA IF NOT EXISTS IDENTIFIER(CONCAT(:catalog,".", :schema,"_serverless"));
```

```python
# Read the json file generated by 01_GenerateTestData

import json

created_tables = json.load(open("./created_tables.json", 'r'))
```

```python
if dbutils.widgets.get("clone_data") == "Yes":
  for table_name in created_tables:
    serverless_table_name = table_name.replace(f".{SCHEMA}.", f".{SCHEMA}_serverless.")
    classic_table_name = table_name.replace(f".{SCHEMA}.", f".{SCHEMA}_classic.")
    spark.sql(f"CREATE OR REPLACE TABLE {serverless_table_name} DEEP CLONE {table_name};")
    spark.sql(f"CREATE OR REPLACE TABLE {classic_table_name} DEEP CLONE {table_name};")
else:
  pass
```
